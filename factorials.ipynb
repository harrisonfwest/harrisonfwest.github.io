```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def factorial(x: int) -> int:
    if x < 2:
        return 1
    else:
        return x * factorial(x - 1)
```

```python
X = np.arange(10)
Y = np.zeros_like(X)
for i in range(len(X)):
    Y[i] = factorial(X[i])
print(X)
print(Y)
```

```python
plt.plot(X, Y)
plt.show()
```
