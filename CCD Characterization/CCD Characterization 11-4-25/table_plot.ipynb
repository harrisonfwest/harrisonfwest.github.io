```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Table.tbl', sep='\t')
df
```

```python
times = [1, 20, 40, 60, 80, 100, 120, 140, 160, 180]
plt.scatter(times, df['Source-Sky'])
plt.xlabel('Integration time (s)')
plt.ylabel('Pixel counts')
plt.title('Pixel counts vs. integration time')
plt.show()
```
